```python
import pandas as pd

# Assuming you uploaded 'co2-emissions-transport.csv'
df = pd.read_csv("co2-emissions-transport.csv")

# Display the first few rows
df.head()
```

```python
# Install any libraries if required (Optional)
# !pip install pandas matplotlib seaborn

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values (if needed)
df.dropna(inplace=True)

# Ensure 'Year' is an integer type and 'CO2 Emissions' is a float
df['Year'] = df['Year'].astype(int)
df['Carbon dioxide emissions from transport'] = df['Carbon dioxide emissions from transport'].astype(float)
```

```python
# Drop the Code column
df = df.drop(columns=['Code'])
```

```python
# Filter for a specific country
country = 'Australia'
country_data = df[df['Entity'] == country]

# Plot CO2 emissions over time
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(country_data['Year'], country_data['Carbon dioxide emissions from transport'], marker='o')
plt.title(f'CO2 Emissions from Transport in {country}')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (tons)')
plt.grid(True)
plt.show()
```

```python
# Filter for multiple countries (e.g., 'United States', 'China', 'India', 'Australia')
countries = ['United States', 'China', 'India','Australia']
filtered_data = df[df['Entity'].isin(countries)]

# Plot CO2 emissions over time for these countries
plt.figure(figsize=(10, 6))
for country in countries:
    country_data = filtered_data[filtered_data['Entity'] == country]
    plt.plot(country_data['Year'], country_data['Carbon dioxide emissions from transport'], marker='o', label=country)

plt.title('CO2 Emissions from Transport in Selected Countries')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (tons)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with missing values in the CO2 emissions column
df_cleaned = df.dropna(subset=['Carbon dioxide emissions from transport'])

# Pivot the data for a heatmap using pivot_table
pivot_data = df_cleaned.pivot_table(values='Carbon dioxide emissions from transport', index='Year', columns='Entity')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of CO2 Emissions from Transport Across Countries')
plt.show()


```

```python
# Customize the heatmap further
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='YlGnBu', linewidths=0.5, annot=True, fmt='.0f')  # 'annot=True' to show the numbers
plt.title('Heatmap of CO2 Emissions from Transport Across Countries')
plt.show()
```

```python
# Get the top 10 emitting countries by total emissions
top_countries = df_cleaned.groupby('Entity')['Carbon dioxide emissions from transport'].sum().nlargest(10).index.tolist()

# Add Australia to the list if it's not already included
if 'Australia' not in top_countries:
    top_countries.append('Australia')

# Filter the data for the selected countries (top 10 + Australia)
top_data = df_cleaned[df_cleaned['Entity'].isin(top_countries)]

# Pivot and plot
pivot_data = top_data.pivot_table(values='Carbon dioxide emissions from transport', index='Year', columns='Entity')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of CO2 Emissions from Transport (Top 10 Countries + Australia)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()

```

```python
# Filter for recent years (e.g., after 2000)
recent_data = df_cleaned[df_cleaned['Year'] >= 2000]

# Pivot and plot
pivot_data = recent_data.pivot_table(values='Carbon dioxide emissions from transport', index='Year', columns='Entity')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of CO2 Emissions from Transport (2000 and later)')
plt.show()
```

```python
# Inspect the dataset for basic info
ev_data.info()

# Clean and select relevant columns
ev_data_cleaned = ev_data[['Model Year', 'Electric Vehicle Type', 'Electric Range']]

# Handle missing values by dropping them (if necessary)
ev_data_cleaned = ev_data_cleaned.dropna()

# Aggregate EV sales by year and type
ev_sales_by_year = ev_data_cleaned.groupby(['Model Year', 'Electric Vehicle Type']).size().reset_index(name='EV_Sales')

# Display the aggregated EV sales data
ev_sales_by_year.head()
```

```python
# Assuming df_cleaned contains the cleaned CO2 emissions data
# Rename 'Model Year' to 'Year' for merging
ev_sales_by_year = ev_sales_by_year.rename(columns={'Model Year': 'Year'})

# Merge the EV sales data with the CO2 emissions data on 'Year'
merged_data = pd.merge(df_cleaned, ev_sales_by_year, how='inner', left_on='Year', right_on='Year')

# Display the merged dataset
merged_data.head()
```

```python
import matplotlib.pyplot as plt

# Example: Filter for a specific country (e.g., United States)
country = 'United States'
country_data = merged_data[merged_data['Entity'] == country]

# Create a dual-axis plot: CO2 emissions vs. EV sales
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot CO2 emissions on the left y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('CO2 Emissions (tons)', color='tab:red')
ax1.plot(country_data['Year'], country_data['Carbon dioxide emissions from transport'], color='tab:red', label='CO2 Emissions')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create another y-axis for EV sales
ax2 = ax1.twinx()
ax2.set_ylabel('EV Sales', color='tab:blue')
ax2.plot(country_data['Year'], country_data['EV_Sales'], color='tab:blue', label='EV Sales')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Add title
plt.title(f'CO2 Emissions vs. EV Sales in {country}')
fig.tight_layout()
plt.show()
```

```python
# Check correlation between EV sales and CO2 emissions
correlation = merged_data['EV_Sales'].corr(merged_data['Carbon dioxide emissions from transport'])
print(f'Correlation between EV Sales and CO2 Emissions: {correlation:.2f}')
```

**Possible Reasons for Low Correlation:**

1. Time Lag Effect:
* The impact of EV adoption on CO2 emissions might not be immediate. CO2 emissions reductions could occur over a longer time period due to the gradual phasing out of internal combustion engine (ICE) vehicles, infrastructure development, and overall EV penetration.

2. Scale of EV Adoption:
* EV sales might still represent a relatively small portion of the total vehicle population in the countries analyzed. Thus, the reduction in emissions may not yet be significant enough to show a strong correlation.

3. Other Factors Influencing CO2 Emissions:
* CO2 emissions from transportation can be influenced by many factors, such as public transportation usage, policy changes, or the economic activity of a region, which may obscure the impact of EV adoption alone.

4. Data Coverage:
* The dataset you’re using may not cover enough time periods or regions to show a strong correlation. For instance, it may not have sufficient data points from regions where EV adoption has been more impactful (e.g., Norway, Netherlands).

**Analyze Impact per Electric Vehicle Type**

```python
# Group EV sales by type and year
ev_type_sales = ev_sales_by_year.groupby(['Year', 'Electric Vehicle Type']).sum().reset_index()

# Merge the CO2 emissions data with EV type sales
merged_ev_type = pd.merge(df_cleaned, ev_type_sales, how='inner', on='Year')

# Plot the comparison of CO2 emissions vs. different EV types
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_ev_type, x='Year', y='Carbon dioxide emissions from transport', hue='Electric Vehicle Type')
plt.title('CO2 Emissions and EV Types Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (tons)')
plt.grid(True)
plt.show()
```

**Predictive Analysis**

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data for regression (using CO2 emissions as the dependent variable and EV sales as the independent variable)
X = merged_data[['EV_Sales']].values
y = merged_data['Carbon dioxide emissions from transport'].values

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the actual vs predicted emissions
plt.figure(figsize=(12, 6))
plt.scatter(merged_data['EV_Sales'], merged_data['Carbon dioxide emissions from transport'], color='blue', label='Actual Emissions')
plt.plot(merged_data['EV_Sales'], y_pred, color='red', label='Predicted Emissions')
plt.title('Predicting CO2 Emissions based on EV Sales')
plt.xlabel('EV Sales')
plt.ylabel('CO2 Emissions (tons)')
plt.legend()
plt.show()
```
