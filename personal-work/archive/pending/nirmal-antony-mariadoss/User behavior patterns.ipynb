# Total EV sales Trend Behavior patterns over time by States

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'D:/1Assignments_T2/Team project (A)/Datasets/sales dataset.xlsx'  
df = pd.read_excel(file_path)

df.columns = df.columns.str.strip().str.replace(' ', '_')

df['Sales_Total'] = pd.to_numeric(df['Sales_Total'], errors='coerce')

sns.set(style='whitegrid')

plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Quarter', y='Sales_Total', hue='States', marker='o')

plt.title('Total EV Sales Trend behavior patterns Over Time by State', fontsize=16)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Sales Total', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Total EV sales Trend Behavior Patterns in Victoria

```python
victoria_data = df[df['States'] == 'Victoria']

plt.figure(figsize=(10, 6))
sns.lineplot(data=victoria_data, x='Quarter', y='Sales_Total', marker='o', color='blue')

plt.title('Total EV Sales Over Time in Victoria', fontsize=16)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Sales Total', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
