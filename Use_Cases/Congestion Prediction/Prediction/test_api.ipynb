```python
#read in EVAT.csv and output predictions to EVAT_predictions.csv for 50 random station_ids
import pandas as pd
import requests
import json   

# Read the CSV file
df = pd.read_csv('EVAT.chargers.csv')    
station_ids = df['_id'].tolist()
#pick out 50 random station_ids
station_ids = pd.Series(station_ids).sample(n=50, random_state=1).tolist()
# Prepare the payload
payload = {
    "station_ids": station_ids
}           
# Make the POST request
response = requests.post('http://localhost:8000/predict/batch', json=payload)    
# Check if the request was successful
if response.status_code == 200: 
    predictions = response.json()
   
    
```
