# Introduction

This notebook contains the pipeline for creating the saved model used for predicting Electric Vehicle Charging Demand (in kWh) by suburb. A postcode and a date can be passed to the model, and it will use a combination of live weather data and data ascertained from the Australian Automobile Association, the CSIRO, and other trusted sources to make predictions.

### Data collection information:

Estimated driving distance/efficiency by vehicle type assumptions
- CSIRO Draft 2026 Forecasting Assumptions Update: https://www.aemo.com.au/consultations/current-and-closed-consultations/draft-2026-forecasting-assumptions-update
    - Draft 2026 Forecasting Assumptions Update
      - Table 4: *Assumed driving distance and efficiency, electric vehicles* for estimated driving distance and efficiency by vehicle type
    - Draft 2026 Forecasting Assumptions Update Workbook - Download from here: https://www.aemo.com.au/consultations/current-and-closed-consultations/draft-2026-forecasting-assumptions-update
        - Contains excel sheets full of data, including total number of different EV types by state

Geographic distribution of vehicle registrations **by postcode** and **vehicle type**
- Data from AAA: https://www.aaa.asn.au/research-data/electric-vehicle/

Public Holidays
- Holidays Python library: https://pypi.org/project/holidays/

Weather API
- Open-Meteo: https://open-meteo.com/


### Pipeline steps:

In this notebook, I will predict demand by date and postcode, using a culmination of the above data.
- Load in vehicle efficiency data and average kms from CSIRO forecast document
- Get numbers of vehicles by postcode
- Estimate demand per postcode based on registrations, vehicle types, vehicle efficiencies, and estimated average distance driven per year
- Forecast daily consumption by taking into account whether a given day is holiday/weekend, and the temperature
- Train, evaluate, and save trained model for API deployment

#### Vehicle Registration Data

Next, I will load in data hosted by the *Australian Automobile Association (AAA)* which can be download from [here](https://www.aaa.asn.au/research-data/electric-vehicle/) on the **Geographic distribution of vehicle registrations**. I downloaded it as .xlsx before converting to .csv.

```python
registration_data = pd.read_csv('EV_Index_Registration_Data_2021-2025.csv', dtype={'Postcode': 'string'}) # postcodes stored as string
registration_data.columns = registration_data.columns.str.strip()

# function for adding the leading 0 to Northern Territory postcodes (e.g. 800 => 0800)
def format_postcode(postcode, state):
    
    if state == 'NT' and len(postcode) == 3:
        # add the missing leading zero for NT postcodes
        return '0' + postcode
    else:
        return postcode

# run function for adding leading 0 to NT postcodes
registration_data['Postcode'] = registration_data.apply(lambda row: format_postcode(row['Postcode'], row['State']), axis=1)

print(f'Shape: {registration_data.shape}')
registration_data.head(6)
```

I removed ICE vehicles from the table as they will not be contributing to electric vehicle charging demand:

#### Number of BEVs PHEVs, and hybrids by category (small, medium, large, truck etc.) and state

From the **AEMO Draft 2026 Forecasting Assumptions Update Detailed EV Workbook**, I exported the *Region*, *Vehicle Category*, and *2025-26* columns in the **BEV_Numbers** table to a new csv.

I did the same thing to the same columns from the **PHEV_Numbers** table, and the **Hybrid_Numbers** table.

I named these files 'BEV_numbers_25-26.csv', 'PHEV_numbers_25-26.csv', and 'hybrid_numbers_25-26.csv' respectively.

```python
BEV_numbers = pd.read_csv('BEV_numbers_25-26.csv')
PHEV_numbers = pd.read_csv('PHEV_numbers_25-26.csv')
hybrid_numbers = pd.read_csv('hybrid_numbers_25-26.csv')

# example of data in tables
BEV_numbers.head()
```

## Data processing

BEV_numbers, PHEV_numbers, and hybrid_numbers all contain values for three separate growth states across each region and vehicle type. I will take the average of these and remove the Scenario column to make the data easier to work with.

```python
# clean the data
BEV_numbers['2025-26'] = BEV_numbers['2025-26'].replace('-', np.nan)
BEV_numbers['2025-26'] = BEV_numbers['2025-26'].astype(str).str.replace(',', '')
BEV_numbers['2025-26'] = pd.to_numeric(BEV_numbers['2025-26'], errors='coerce')

# average the counts across each scenario
BEV_numbers_cleaned = BEV_numbers.groupby(['Region', 'Vehicle Category'])['2025-26'].mean().reset_index()

# fill NaN values with 0 and convert to int
BEV_numbers_cleaned['2025-26'] = BEV_numbers_cleaned['2025-26'].fillna(0).round().astype(int)

BEV_numbers_cleaned.head()
```

```python
# clean the data
PHEV_numbers['2025-26'] = PHEV_numbers['2025-26'].replace('-', np.nan)
PHEV_numbers['2025-26'] = PHEV_numbers['2025-26'].astype(str).str.replace(',', '')
PHEV_numbers['2025-26'] = pd.to_numeric(PHEV_numbers['2025-26'], errors='coerce')

# average the counts across each scenario
PHEV_numbers_cleaned = PHEV_numbers.groupby(['Region', 'Vehicle Category'])['2025-26'].mean().reset_index()

# fill NaN values with 0 and convert to int
PHEV_numbers_cleaned['2025-26'] = PHEV_numbers_cleaned['2025-26'].fillna(0).round().astype(int)

PHEV_numbers_cleaned.head()
```

```python
# clean the data
hybrid_numbers['2025-26'] = hybrid_numbers['2025-26'].replace('-', np.nan)
hybrid_numbers['2025-26'] = hybrid_numbers['2025-26'].astype(str).str.replace(',', '')
hybrid_numbers['2025-26'] = pd.to_numeric(hybrid_numbers['2025-26'], errors='coerce')

# average the counts across each scenario
hybrid_numbers_cleaned = hybrid_numbers.groupby(['Region', 'Vehicle Category'])['2025-26'].mean().reset_index()

# fill NaN values with 0 and convert to int
hybrid_numbers_cleaned['2025-26'] = hybrid_numbers_cleaned['2025-26'].fillna(0).round().astype(int)

hybrid_numbers_cleaned.head()
```

Since we will be predicting EV charging demand using the registrations of light electric vehicles, I will remove all heavy vehicle records from the BEV, PHEV, and hybrid vehicle tables.

```python
heavy_vehicles = ['Articulated Truck', 'Bus', 'Rigid Truck']

# remove all heavy vehicle records
light_BEVs = BEV_numbers_cleaned[~BEV_numbers_cleaned['Vehicle Category'].isin(heavy_vehicles)].copy()
light_PHEVs = PHEV_numbers_cleaned[~PHEV_numbers_cleaned['Vehicle Category'].isin(heavy_vehicles)].copy()
light_hybrids = hybrid_numbers_cleaned[~hybrid_numbers_cleaned['Vehicle Category'].isin(heavy_vehicles)].copy()
```

I'll now join the tables on *Region* and *Vehicle Category*.

```python
# rename columns to make merging easier
bev_df = light_BEVs.rename(columns={'2025-26': 'BEV'})
phev_df = light_PHEVs.rename(columns={'2025-26': 'PHEV'})
hybrid_df = light_hybrids.rename(columns={'2025-26': 'Hybrid'})

# merge the bev and phev tables
merged_df = pd.merge(bev_df, phev_df, on=['Region', 'Vehicle Category'], how='outer')

# merge the result with the hybrid table
light_EVs = pd.merge(merged_df, hybrid_df, on=['Region', 'Vehicle Category'], how='outer')

light_EVs.head()
```

Some of the values for Region are "WEM". "WEM" stands for the Wholesale Electricity Market, which is the electricity grid covering Western Australia (WA). I'll map WEM to WA so we can merge with ev_registration_data later.

```python
# replace WEM with WA for mapping to ev_registration_data
light_EVs['Region'] = light_EVs['Region'].replace('WEM', 'WA')
```

There is also no data for the ACT in light_EVs, so since it's within NSW and is part of the same electricity grid (the NEM), the most logical thing to do is to use the NSW proportions for the ACT.

```python
# create data for ACT by copying NSW data
act_data = light_EVs[light_EVs['Region'] == 'NSW'].copy()
act_data['Region'] = 'ACT'

# add new ACT data to light_EVs
light_EVs = pd.concat([light_EVs, act_data], ignore_index=True)

# check unique region values
light_EVs['Region'].unique()
```

In the **ev_registration_data** table, we have no way of differentiating between the number of hybrids and PHEVs, as their counts are combined. This is a problem, as hybrids do not contribute to EV charger demand as they self-charge using their engines, so we need to be able to separate the two.

I will use the above CSIRO prediction data that we just cleaned above, which is broken down into vehicle type and by state, to estimate the number of PHEVs and hybrids in each state.

Since the number of records in the ev_registration_data dataset and the BEV/PHEV/hybrid datasets from the CSIRO Workbook are different, I will calculate the ratio of hybrids to PHEVs, and use that value to estimate the number of PHEVs in each postcode in ev_registration_data.

```python
# get ratio of PHEVs to Hybrids
light_EVs['PHEV-Hybrid Ratio'] = light_EVs['PHEV'] / (light_EVs['PHEV'] + light_EVs['Hybrid'])

# replace NaN values (where 0 was divided by 0) with 0
light_EVs['PHEV-Hybrid Ratio'] = light_EVs['PHEV-Hybrid Ratio'].fillna(0)

light_EVs.head(10)
```

Next we create the PHEV vs Hybrid proportions by grouping by state and summing the PHEV and Hybrid counts, and then dividing the number of PHEVs by the total.

```python
state_level_counts = light_EVs.groupby('Region')[['PHEV', 'Hybrid']].sum()
state_level_counts['State_PHEV_Proportion'] = state_level_counts['PHEV'] / (state_level_counts['PHEV'] + state_level_counts['Hybrid'])

# save the proportions to a series for easier mapping later
state_phev_proportion_map = state_level_counts['State_PHEV_Proportion']
print(state_phev_proportion_map)
state_level_counts
```

Next we create the Vehicle Category proportions map:

```python
# calculate the proportion of each vehicle category within its fuel type and state
light_EVs['BEV_Cat_Proportion'] = (light_EVs['BEV'] / light_EVs.groupby('Region')['BEV'].transform('sum')).fillna(0)
light_EVs['PHEV_Cat_Proportion'] = (light_EVs['PHEV'] / light_EVs.groupby('Region')['PHEV'].transform('sum')).fillna(0)

# handle any 0/0 divisions
light_EVs.fillna(0, inplace=True)

# create a clean lookup table with only necessary columns
category_proportions_map = light_EVs[['Region', 'Vehicle Category', 'BEV_Cat_Proportion', 'PHEV_Cat_Proportion']]

category_proportions_map.head()
```

Since we have no data for NT from the CSIRO workbook for calculating fuel type proportions, we can use the national average as a substitute.

```python
national_phev_proportion = light_EVs['PHEV'].sum() / (light_EVs['PHEV'].sum() + light_EVs['Hybrid'].sum())
```

We will also calculate the national vehicle category proportions for use in NT postcodes.

```python
# national category proportions
national_cat_proportions = light_EVs.groupby('Vehicle Category')[['BEV', 'PHEV']].sum()
national_cat_proportions['BEV_Cat_Proportion'] = national_cat_proportions['BEV'] / national_cat_proportions['BEV'].sum()
national_cat_proportions['PHEV_Cat_Proportion'] = national_cat_proportions['PHEV'] / national_cat_proportions['PHEV'].sum()

# create mappings as Series from national_cat_proportions
national_bev_cat_map = national_cat_proportions['BEV_Cat_Proportion']
national_phev_cat_map = national_cat_proportions['PHEV_Cat_Proportion']
```

Now I prepare ev_registration_data by creating a new dataframe that only contains the most recent data, and split it into two new tables (bev_registrations and phev_hybrid_registrations) for processing.

I perform a cross join to ensure every postcode has a row for every vehicle category.

```python
# create a list of all unique light vehicle categories
all_vehicle_categories = light_EVs['Vehicle Category'].unique()
categories_df = pd.DataFrame({'Vehicle Category': all_vehicle_categories})

# prepare latest registrations
latest_registrations = ev_registration_data[['Postcode', 'State', 'Fuel Type', '31-Jan-2025']].rename(
    columns={'31-Jan-2025': 'Count'}
)
bev_registrations = latest_registrations[latest_registrations['Fuel Type'] == 'BEV'].copy()
phev_hybrid_registrations = latest_registrations[latest_registrations['Fuel Type'] == 'Hybrid/PHEV'].copy()

# perform cross join to create all combinations so every postcode will have a row for every vehicle category
bev_registrations_expanded = pd.merge(bev_registrations, categories_df, how='cross')
phev_hybrid_registrations_expanded = pd.merge(phev_hybrid_registrations, categories_df, how='cross')
```

To disaggregate the BEV data, we perform a left merge on both the State/Region and Vehicle Category features. The NT postcodes will have NaN values, so we fill them using the national average (as this data wasn't available for the NT).

Finally, we calculate the estimated number of BEVs per category per postcode by multiplying by the total count of EVs from the original registrations dataset with the BEV category proportion. 

```python
# merge on both State/Region and Vehicle Category
final_bev_data = pd.merge(
    bev_registrations_expanded,
    category_proportions_map,
    how='left',
    left_on=['State', 'Vehicle Category'],
    right_on=['Region', 'Vehicle Category']
)

# fill NaN proportions using national average map (for NT)
final_bev_data['BEV_Cat_Proportion'] = final_bev_data['BEV_Cat_Proportion'].fillna(
    final_bev_data['Vehicle Category'].map(national_bev_cat_map)
)

# calculate estimated counts
final_bev_data['Estimated Count BEVs'] = final_bev_data['Count'] * final_bev_data['BEV_Cat_Proportion']
```

We perform the same process for PHEV data, but instead we start by estimating the total number of PHEVs by removing the estimated number of Hybrids in the combined count (as we don't want hybrid vehicles in our calculations as they aren't charged by the grid but by their own engine).

```python
# add the total PHEV count to the expanded DataFrame
phev_hybrid_registrations_expanded['State_PHEV_Proportion'] = phev_hybrid_registrations_expanded['State'].map(state_phev_proportion_map).fillna(national_phev_proportion)
phev_hybrid_registrations_expanded['Total_PHEVs_in_Postcode'] = phev_hybrid_registrations_expanded['Count'] * phev_hybrid_registrations_expanded['State_PHEV_Proportion']

# merge on both State/Region and Vehicle Category
final_phev_data = pd.merge(
    phev_hybrid_registrations_expanded,
    category_proportions_map,
    how='left',
    left_on=['State', 'Vehicle Category'],
    right_on=['Region', 'Vehicle Category']
)

# fill NaN proportions using national average map (for NT)
final_phev_data['PHEV_Cat_Proportion'] = final_phev_data['PHEV_Cat_Proportion'].fillna(
    final_phev_data['Vehicle Category'].map(national_phev_cat_map)
)

# calculate estimated counts
final_phev_data['Estimated Count PHEVs'] = final_phev_data['Total_PHEVs_in_Postcode'] * final_phev_data['PHEV_Cat_Proportion']
```

We select only the 'Postcode', 'State', 'Vehicle Category', and 'Estimated Count BEVs/PHEVs' columns, then merge on 'Postcode', 'State', and 'Vehicle Category'.

We replace any remaining NaN values with 0 and round the counts to integers.

```python
# select only necessary columns
final_bev_subset = final_bev_data[['Postcode', 'State', 'Vehicle Category', 'Estimated Count BEVs']]
final_phev_subset = final_phev_data[['Postcode', 'State', 'Vehicle Category', 'Estimated Count PHEVs']]

# merge the BEV and PHEV dataframes
final_disaggregated_data = pd.merge(
    final_bev_subset,
    final_phev_subset,
    on=['Postcode', 'State', 'Vehicle Category'],
    how='outer'
)

# replace NaN with 0
final_disaggregated_data.fillna(0, inplace=True)

# convert counts to integer by rounding
final_disaggregated_data[['Estimated Count BEVs', 'Estimated Count PHEVs']] = final_disaggregated_data[['Estimated Count BEVs', 'Estimated Count PHEVs']].round().astype(int)

final_disaggregated_data.head()
```

**Joining the efficiency_data and final_disaggregated_data tables:**

I rename the vehicle_type feature in efficiency_data to Vehicle Category to allow easier joining. This merge results in a table containing the estimated yearly driving distance and vehicle efficiency values for each vehicle category as provided by the CSIRO.

Efficiency is in kWh/km.

```python
efficiency_data = efficiency_data.rename(columns={"vehicle_type": "Vehicle Category"})

postcode_demand_data = pd.merge(
    final_disaggregated_data,
    efficiency_data,
    on='Vehicle Category',
    how='inner'
)

postcode_demand_data.head()
```

The current table contains static information. To predict daily demand, we need to transform this into a time-series dataset. The core formula we'll be working with is:

    Daily Demand (kWh) = Number of Vehicles * Daily Driving Distance * Vehicle Efficiency (kWh/km)

The next step is to calculate a baseline daily demand, then build a model that adjusts this baseline based on other factors.

**Calculating the baseline daily demand:**

We start by calculating the average daily demand in kWh for each vehicle category in each postcode.

```python
# calculate the total number of vehicles that require grid charging
postcode_demand_data['total_vehicles'] = postcode_demand_data['Estimated Count BEVs'] + postcode_demand_data['Estimated Count PHEVs']

# convert yearly driving distance to daily
postcode_demand_data['daily_driving_distance'] = postcode_demand_data['yearly_driving_distance'] / 365

# calculate the baseline daily kWh demand for each vehicle category in each postcode
postcode_demand_data['baseline_daily_kwh'] = (
    postcode_demand_data['total_vehicles'] *
    postcode_demand_data['daily_driving_distance'] *
    postcode_demand_data['efficiency']
)

# table with one row per postcode with its baseline daily demand
postcode_baseline = postcode_demand_data.groupby(['Postcode', 'State'])['baseline_daily_kwh'].sum().reset_index()

postcode_baseline.head()
```

**Creating the synthetic training dataset:**

We don't have actual historical demand data, so we have to be creative with making our training data based on our assumptions. We are going to create a high-quality, synthetic target variable to train our model on.

We know the following:
- Demand is higher when it's cold (EV batteries are less efficient).
- Demand is lower on public holidays.
- Weekend demand patterns are different from weekday patterns.

We can use this information to generate a realistic historical dataset. Then, we can train a machine learning model to learn these rules automatically from the data.

```python
# create a date range for the training data
end_date = datetime.now() + timedelta(days=-1)  # include up to yesterday
dates = pd.date_range(start='2023-01-01', end=end_date, freq='D')

# store dates in df
dates_df = pd.DataFrame({'Date': dates})

# get a unique list of postcodes
postcodes_df = postcode_baseline[['Postcode', 'State']].drop_duplicates()

# create the time series framework with a cross join
training_df = pd.merge(postcodes_df, dates_df, how='cross')

# merge baseline demand onto this skeleton
training_df = pd.merge(
    training_df,
    postcode_baseline,
    on=['Postcode', 'State'],
    how='left'
)

print(f'Shape: {training_df.shape}')
training_df.head()
```

Next, we can add **basic temporal features** (day of week, etc.) by using .dt in pandas to extract useful features from the Date column.

```python
# ensure Date column is a datetime object
training_df['Date'] = pd.to_datetime(training_df['Date'])

# Day and Week features
training_df['day_of_week'] = training_df['Date'].dt.dayofweek  # Monday=0, Sunday=6
training_df['is_weekend'] = (training_df['day_of_week'] >= 5).astype(int) # 1 for weekend, 0 for weekday

# Month and Year features
training_df['month'] = training_df['Date'].dt.month
training_df['day_of_year'] = training_df['Date'].dt.dayofyear
training_df['quarter'] = training_df['Date'].dt.quarter
training_df['year'] = training_df['Date'].dt.year

training_df.head()
```

Now we can add **public holiday features**. The following cell uses the *holidays* library imported at the start of the notebook.

```python
# get the unique years and states from our DataFrame
years = training_df['year'].unique()
states = training_df['State'].unique()

# create a dictionary to hold holiday objects for each state for faster lookups
state_holidays = {
    state: holidays.AU(subdiv=state, years=years) for state in states if state in holidays.AU.subdivisions
}

# function to check if a given date is a holiday in its specific state
def is_state_holiday(row):
    state = row['State']
    date = row['Date']
    # Check if the state exists in our holiday dictionary and if the date is a holiday
    if state in state_holidays and date in state_holidays[state]:
        return 1
    return 0

# apply this function to every row to create the 'is_holiday' column
print("\nCalculating holidays... (This may take a moment)")
training_df['is_holiday'] = training_df.apply(is_state_holiday, axis=1)
print("Holiday calculation complete.")

# print number of holidays found
print(f"\nTotal holiday days found: {training_df['is_holiday'].sum()}")

print("\nCurrent state  of training dataframe:")
print(training_df.sample(n=10, random_state=42)) # Show a random sample
```



We all know that there's less traffic on weekends based on personal experience, however, actual data on this fact is very hard to come by. The closest analogy to Australian weekend traffic I could find was New Zealand weekend traffic data from the Australasian Transport Research Forum, which states:

> Unpublished vehicle count data collected for one week periods in 2001 for major urban arterials and state highways in Wellington region shows that mean **Saturday traffic volumes are about 90%** of the average weekday traffic flows, and in several cases (in the CBD and the suburbs of Newtown and Karori) are exceeding weekday traffic flows. **Sunday vehicle counts are much lower, with flows approximately 73%** - https://australasiantransportresearchforum.org.au/wp-content/uploads/2022/03/2003_OFallon_Sullivan_c.pdf

I will use these values when estimating distance travelled on weekends.

Additionally, a VicRoads study found about a **15% reduction in peak-hour traffic during holidays** compared to regular days. I'll flatten this to a 15% overall amount of less km travelled on public holidays.

**Estimating Electric Vehicle efficiency based on temperature:**

As can be seen from the following visualisation, EV batteries perform worse in colder weather, and their range drops. This means on colder days we will use more kWh per km travelled, and demand will go up.

 Source: https://www.geotab.com/blog/ev-range/

I will take the points from the above graph, and use them as a basis for interpolating a demand multiplier based on a given temperature.

```python
# manually extract (Temp, Range %) points from the above graph
temps_c =        [-20,-10,  0,  10,  20,  30, 35, 40]
range_percents = [45,  60, 80, 100, 115, 105, 90, 80]

# define the peak efficiency from the data
peak_range_percent = 115.0

# calculate the demand multiplier for each point
demand_multipliers = [peak_range_percent / r for r in range_percents]

# create a DataFrame to store values
efficiency_lookup = pd.DataFrame({
    'Temp_C': temps_c,
    'Range_Percent': range_percents,
    'Demand_Multiplier': demand_multipliers
})

efficiency_lookup
```

We create a new function using numpy.interp that gets accurate multipliers using interpolation:

```python
def get_temperature_multiplier_interp(temp):
    """
    Calculates demand multiplier for a given temperature using linear interpolation.
    """
    
    # handles temperatures outside the range by returning the value of the nearest endpoint.
    return np.interp(temp, temps_c, demand_multipliers) # temps_c and demand_multipliers defined in above code block
```

Convert postcodes to coordinates for passing to API.

```python
import pgeocode

# get a unique list of postcodes from our training data
unique_postcodes = training_df['Postcode'].unique()

# initialize the geocoder for Australia
nomi = pgeocode.Nominatim('au')

# create a dictionary to store {postcode: (lat, lon)}
postcode_coords = {}
print("Converting postcodes to coordinates...")
for pc in unique_postcodes:
    # query postcode
    location = nomi.query_postal_code(pc)
    # check if we got a valid location and store it
    if pd.notna(location.latitude):
        postcode_coords[pc] = (location.latitude, location.longitude)

print(f"Successfully converted {len(postcode_coords)} postcodes.")
```

Now we will download a master list of ALL weather stations worldwide from NOAA, then filter down to keep only Australia stations.

```python
# code for getting station metadata and mapping postcodes
stations_url = "https://www.ncei.noaa.gov/pub/data/ghcn/daily/ghcnd-stations.txt"
col_widths = [11, 9, 10, 7, 3, 31, 4, 4, 6]
col_names = ['ID', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'STATE', 'NAME', 'GSN_FLAG', 'HCN_CRN_FLAG', 'WMO_ID']
stations_df = pd.read_fwf(stations_url, widths=col_widths, names=col_names)
au_stations = stations_df[stations_df['ID'].str.startswith('ASN')].copy() # au_stations contains all Australian weather stations and their locations
```

Now, we can extract the latitude/longitude coordinates of all Australian weather stations, and create a KD-Tree data structure (*cKDTree*) for efficient spatial indexing and nearest-neighbor searches (for finding nearest weather stations to postcode coordinates).

```python
from scipy.spatial import cKDTree

# extract just the latitude and longitude coordinates of all Australian weather stations
station_coords = au_stations[['LATITUDE', 'LONGITUDE']].values

# create KD-Tree data structure for efficient spatial indexing and nearest-neighbor searches
kdtree = cKDTree(station_coords)
```

We'll use the following code block to **fetch historical weather data**.

For postcodes inside the Northern Territory, they will all unfortunately have to get their historical weather data from the one weather station (Tindal RAAF, ASN00014932), as none of the other stations in the NT had temperature data, only precipitation.

We're also unable to get historical weather data information from the weather stations in Canberra using the www.ncei.noaa.gov data. Instead, I will use the closest weather station to Canberra (Wagga Wagga AMO, ASN00072150) inside NSW for Canberra's weather data.

This is a limitation with the free data from www.ncei.noaa.gov data; I'm sure there are paid options that contain more accurate temperature data.

**NOTE:** It will take quite a few minutes to run if there is no existing *historical_weather.csv* file. If you already have the file, but want to redownload the data, you will have to delete it first.

```python
import requests

postcode_to_station_map = {}

# special stations for problem areas
working_nt_station = 'ASN00014932' # the one NT station that has temperature data
working_act_station = 'ASN00072150'  # closest working NSW station

print('Mapping postcodes to their nearest weather station...')

# mapping postcodes to nearest weather stations
for postcode, (lat, lon) in tqdm(postcode_coords.items()):
    
    # special handling for NT postcodes
    if str(postcode).startswith('0') and len(postcode) == 4:  # NT postcodes (e.g. 0800)
        # force all NT postcodes to use the one working station
        postcode_to_station_map[postcode] = working_nt_station

    # Special handling for ACT postcodes (2600-2618)
    elif postcode.startswith('2') and 2600 <= int(postcode) <= 2618:
        postcode_to_station_map[postcode] = working_act_station
        
    else:
        # for all other states, use the nearest station
        distance, index = kdtree.query([lat, lon], k=1)
        nearest_station_id = au_stations.iloc[index]['ID']
        postcode_to_station_map[postcode] = nearest_station_id

print(f"Successfully mapped {len(postcode_to_station_map)} postcodes to weather stations.")

# --- download/load weather data ---

# define filename
weather_filename = 'historical_weather.csv'

# download weather data if no historical_weather.csv found
if not os.path.exists(weather_filename):
    print("Weather data file not found. Starting download process...")
    
    years_to_download = training_df['year'].unique()
    all_years_weather = []

    # download and processing loop
    for year in years_to_download:
        print(f"Downloading and processing bulk weather data for {year}...")
        data_url = f"https://www.ncei.noaa.gov/pub/data/ghcn/daily/by_year/{year}.csv.gz"
        try:
            daily_df_year = pd.read_csv(data_url, compression='gzip', header=None, names=['ID', 'DATE', 'ELEMENT', 'VALUE', 'M_FLAG', 'Q_FLAG', 'S_FLAG', 'OBS_TIME'])
            needed_stations = list(postcode_to_station_map.values())
            daily_df_year = daily_df_year[daily_df_year['ID'].isin(needed_stations) & daily_df_year['ELEMENT'].isin(['TMAX', 'TMIN'])]
            weather_data_year = daily_df_year.pivot_table(index=['ID', 'DATE'], columns='ELEMENT', values='VALUE').reset_index()
            weather_data_year['temperature'] = ((weather_data_year['TMAX'] + weather_data_year['TMIN']) / 2) / 10.0
            weather_data_year['DATE'] = pd.to_datetime(weather_data_year['DATE'], format='%Y%m%d')
            weather_data_year = weather_data_year[['ID', 'DATE', 'temperature']].rename(columns={'ID': 'station_id', 'DATE': 'Date'})
            all_years_weather.append(weather_data_year)
            print(f"Successfully processed {year}.")
        except Exception as e:
            print(f"Could not download or process data for {year}. Error: {e}")

    # create the dataframe first
    weather_df = pd.concat(all_years_weather, ignore_index=True)
    
    # save it to CSV - 'index=False' avoids extra column
    print(f"Saving processed weather data to '{weather_filename}'...")
    weather_df.to_csv(weather_filename, index=False)
    print("Save complete.")

else:
    print(f"Found existing weather data file: '{weather_filename}'. Loading from disk.")
    # load weather data file
    weather_df = pd.read_csv(weather_filename)
    
    # convert the 'Date' column to datetime
    weather_df['Date'] = pd.to_datetime(weather_df['Date'])

print("\nWeather data ready.")
print(weather_df.info())
```

Now we will merge our weather data into our training dataframe. We will impute missing temperature values in three steps:
1. For each station, if there's a missing day between two days with data, it estimates the temperature by drawing a straight line between them.
2. For each state and each specific date, calculates the average temperature across all stations in that state to fill remaining missing values.
3. Calculate the overall average temperature for each state across ALL dates and fill any remaining missing values.

```python
# connect Postcodes to Weather Stations
training_df['station_id'] = training_df['Postcode'].map(postcode_to_station_map)

# merge Weather Data
training_df = pd.merge(training_df, weather_df, on=['station_id', 'Date'], how='left')

print(f"Found {training_df['temperature'].isna().sum()} missing temperature values before imputation.")

# fill missing temperature values by using previous and next day within station (if temps for those days exist)
training_df = training_df.sort_values(['station_id', 'Date'])
training_df['temperature'] = training_df.groupby('station_id')['temperature'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

# for remaining missing temperature values, fill using average temperature across all stations in that state on that date
state_daily_mean_map = training_df.groupby(['State', 'Date'])['temperature'].transform('mean')
training_df['temperature'] = training_df['temperature'].fillna(state_daily_mean_map)

# for any final missing temperature values, fill using average temperature across all stations in that state across whole time period
state_overall_mean_map = training_df.groupby('State')['temperature'].transform('mean')
training_df['temperature'] = training_df['temperature'].fillna(state_overall_mean_map)

print(f"Imputation complete. {training_df['temperature'].isna().sum()} missing values remain.")
```

The final step in creating our training data is to apply our weekend, holiday, and temperature multipliers to the baseline daily demand values. Following this, we can begin training and evaluating models.

```python
# define weekend/holiday multipliers
saturday_multiplier = 0.90
sunday_multiplier = 0.73
holiday_multiplier = 0.85

# start with baseline demand
training_df['simulated_daily_kwh'] = training_df['baseline_daily_kwh']

# apply driving distance multipliers
training_df.loc[training_df['day_of_week'] == 5, 'simulated_daily_kwh'] *= saturday_multiplier
training_df.loc[training_df['day_of_week'] == 6, 'simulated_daily_kwh'] *= sunday_multiplier
training_df.loc[training_df['is_holiday'] == 1, 'simulated_daily_kwh'] *= holiday_multiplier

# apply temperature multiplier
temp_multipliers = training_df['temperature'].apply(get_temperature_multiplier_interp)
training_df['simulated_daily_kwh'] *= temp_multipliers

print("\nExample data from final training dataframe:\n")
print(training_df[['Date', 'temperature', 'baseline_daily_kwh', 'simulated_daily_kwh']].sample(n=10, random_state=42))
```

The following function is used to get temperature forecasts from https://open-meteo.com/.

```python
def get_weather_forecast(lat, lon, target_date):
    """
    Fetches the forecasted mean temperature for a single location and date
    by calling the Open-Meteo API directly with the requests library.
    """
    # calculate how many days ahead the target date is
    days_ahead = (target_date.date() - datetime.now().date()).days + 1
    if days_ahead < 1 or days_ahead > 16:
        raise ValueError("Can only fetch forecasts for the next 16 days.")

    # url for API
    base_url = "https://api.open-meteo.com/v1/forecast"
    
    # parameters are added directly to the URL
    params = {
        "latitude": lat,
        "longitude": lon,
        "daily": "temperature_2m_mean",
        "forecast_days": days_ahead,
        "timezone": "Australia/Sydney"
    }

    try:
        # HTTP GET request.
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # raises an error for bad responses
        
        # parse the JSON response.
        data = response.json()
        
        # convert results into a pandas DataFrame.
        forecast_df = pd.DataFrame(data['daily'])
        forecast_df['time'] = pd.to_datetime(forecast_df['time'])
        
        # finds the temperature for the specific target date.
        target_date_str = target_date.strftime('%Y-%m-%d')
        daily_temp = forecast_df[forecast_df['time'] == target_date_str]['temperature_2m_mean'].iloc[0]
        
        return float(daily_temp)

    except requests.exceptions.RequestException as e:
        print(f"HTTP Request Error: {e}")
        return None
    except (KeyError, IndexError) as e:
        print(f"Failed to parse weather data response. Error: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred in get_weather_forecast: {e}")
        return None
```

The following two functions generate the data for predictions based on the given date, postcode, and data received by the weather API (from the date and postcode).

```python
def create_prediction_features(postcode, target_date):
    """
    Creates a single row of features for one postcode and one date,
    matching the LightGBM model's training data.
    """
    # get static data from postcode_baseline
    baseline_info = postcode_baseline[postcode_baseline['Postcode'] == postcode]
    if baseline_info.empty:
        raise ValueError(f"Postcode '{postcode}' not found in baseline data.")
    
    baseline_kwh = baseline_info['baseline_daily_kwh'].iloc[0]
    state = baseline_info['State'].iloc[0]
    
    # generate date-based features
    day_of_week = target_date.weekday()
    month = target_date.month
    day_of_year = target_date.timetuple().tm_yday
    year = target_date.year
    is_holiday = 1 if target_date in au_holidays else 0
    
    # fetch weather data
    if postcode not in postcode_coords:
        raise ValueError(f"Coordinates for postcode '{postcode}' not found.")
    lat, lon = postcode_coords[postcode]
    
    temperature = get_weather_forecast(lat, lon, target_date)
    if temperature is None:
        temperature = 20.0 
        print(f"Warning: Weather forecast failed for {postcode}. Using default temp of {temperature}C.")

    # assemble feature dictionary to match training data
    feature_dict = {
        'Postcode': postcode,
        'State': state,
        'baseline_daily_kwh': baseline_kwh,
        'day_of_week': day_of_week,
        'is_weekend': 1 if day_of_week >= 5 else 0,
        'month': month,
        'day_of_year': day_of_year,
        'quarter': (month - 1) // 3 + 1,
        'year': year,
        'is_holiday': is_holiday,
        'temperature': temperature
    }
    
    # create df with correct dtypes and column order
    pred_df = pd.DataFrame([feature_dict])
    
    # convert Postcode and State to the same category type as the training data
    pred_df['Postcode'] = pred_df['Postcode'].astype('category')
    pred_df['State'] = pred_df['State'].astype('category')
    
    # use the feature_columns list to ensure the order is identical to the training data.
    return pred_df[feature_columns]


def predict_demand(postcode, date_str):
    """
    High-level function to predict demand for a given postcode and date string.
    """
    try:
        target_date = datetime.strptime(date_str, '%Y-%m-%d')
        
        # create feature vector for the new data point
        X_pred = create_prediction_features(postcode, target_date)
        
        # make prediction
        prediction = best_model.predict(X_pred)
        
        return {
            "postcode": postcode,
            "date": date_str,
            "predicted_demand_kwh": round(prediction[0], 2),
            "status": "success"
        }
    except Exception as e:
        return {
            "postcode": postcode,
            "date": date_str,
            "error": str(e),
            "status": "error"
        }
```
